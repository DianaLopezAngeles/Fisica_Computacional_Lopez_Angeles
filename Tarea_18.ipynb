## Sistemas dinamicos discretos

```julia
using PyPlot
```

```julia
function diagrama_de_bifurcacion(x,r,n)
    for i in 1:n                               #Un ciclo for que se va a hacer n veces para la curva logistica.
        x = r*x*(1-x)                          #Curva logistica.
    end
    return x
    end;

function exponente_Lyapunov(x,n,r)
    l=0    
for i in 1:n                                  #Un ciclo for que se va a hacer n veces para la curva logistica.
    x = r*x*(1-x)                             #Curva logistica.
    l = l+log(abs(r*(1-2*x)))                 #Exponente de Lyaponuv
    end
    return l/n
    end;
```

```julia
#Aqui vamos a hacer todas las iteraciones para poder graficar así como los intervalos donde queremos que este la r y la x.
las_equis=linspace(0,1,1000)
las_erres=linspace(1,4,1000)
m2=linspace(1,3.9,1000)
lista_equis =[]
lista_erres =[]
for i in 1:length(las_erres)
    push!(lista_equis,diagrama_de_bifurcacion.(las_equis,las_erres[i],1000))
    push!(lista_erres,ones(length(las_equis))*las_erres[i])
    end;
```

```julia
plot(lista_erres,lista_equis,"o",markersize=0.005);
title("Diagrama de bifurcacion")
xlabel("r")
ylabel("x")
legend();
```

```julia
plot(las_erres,exponente_Lyapunov.(0.5,1000,las_erres),"o",markersize=0.5)
title("Exponente de Lyapunov")
legend();
```
