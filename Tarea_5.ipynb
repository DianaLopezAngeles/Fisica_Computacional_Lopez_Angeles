### Método de Integración de Riemann
-----

Una partición $P={a=p_0,p_1,...,p_n=b}$ del intervalo $[a,b]$ forma un mosaico de intervalos cerrados ${P_1,...,P_n}$ que cubre al intervalo $[a,b]$. La medida del i-esimo intervalo es:

$$
\lambda(P_i)=p_i-p_{i-1}
$$

Las sumas inferiores y superiores:
$$
L(f,P)=\displaystyle\sum_{i=1}^n \lambda(P_i)l(f,P_i)
$$
$$
U(f,P)=\displaystyle\sum_{i=1}^n \lambda(P_i)u(f,P_i)
$$

donde:
$$
l(f,A)=inf\left\{f(x):x\in\mathbb{A}\right\}
$$
$$
u(f,A)=sup\left\{f(x):x\in\mathbb{A}\right\}
$$

Una función $f$ acotada en un intervalo $[a,b]$ es integrable si
$$
sup\left\{L(f,P):P\in\mathbb{\mathcal{P}}_{[a,b]}\right\}=\int_{[a,b]}f=inf\left\{U(f,P):P\in\mathbb{\mathcal{P}}_{[a,b]}\right\}
$$

### Regla del Trapecio
----
La expresión que aproxima el valor de la integral se conoce como regla del trapecio, porque geométricamente se puede interpretar que se aproxima el área bajo la curva por el área bajo un polinomio de grado uno $P_1(x)$ y la figura que resulta es un trapecio tal como se observa en la figura.

![Ilustración de la regla del Trapecio](Regla_del_trapecio.jpg "Ilustración de la regla del Trapecio")

Consideremos una funcion $f$ en el intervalo $[a,b]$, con los puntos $(a,f(a))$ y $(b,f(b))$ se construye un polinomio de Lagrange de grado uno.

$$
f(x)=P_1(x)+E
$$

Con E el error en la aproximación. Luego:

$$
f(x)=\frac{(x-a)}{a-b}f(a)+\frac{(x-a)}{b-a}f(b)+\frac{f'''(z)}{2}(x-a)(x-b)
$$

Al integrar $f(x)$ obtenemos:

$$
\int_{a}^bf(x)dx\approx\frac{(b-a)}{2}[f(b)+f(a)]
$$

Con $h=b-a$

$$
\int_{a}^bf(x)dx\approx\frac{h}{2}[f(b)+f(a)]
$$

El error para esta aproximación está dado por:

$$
E= \left | \frac{h^3}{12}f''(z)\right | 
$$

con $a<z<b$.

### Regla de Simpson
-------------
Una forma de mejorar la aproximación de una integral es con una mejor aproximación para el integrando ƒ(x). Esto se puede lograr con un polinomio de grado 2.

![Ilustración de la regla de Simpson](Simpson_integral.jpg "Ilustración de la regla de Simpson")

Considerece la ecuacion $f(x)$ en el intervalo $[a,b]$ y además:

$$
x_0=a
$$

$$
x_1=x_0+h
$$

$$
x_2=b
$$

Donde:

$$
h=\frac{b-a}{2}
$$

Con los puntos $(x_0,f(x_0)),(x_1,f(x_1))$ y $(x_2,f(x_2))$ se construye el polinomio de Lagrange de grado 2.

$$
P_2(x)=\frac{(x-x_1)(x-x_2)}{(x_0-x_1)(x_0-x_2)}f(x_0)+\frac{(x-x_0)(x-x_2)}{(x_1-x_0)(x_1-x_2)}f(x_1)+\frac{(x-x_0)(x-x_1)}{(x_2-x_0)(x_2-x_1)}f(x_2)
$$

Ahora:

$$
\int_{x_0}^{x_2}f(x)dx\approx\int_{x_0}^{x_2}P_2(x)dx
$$

La integral del polinomio se resuelve por partes, omitiremos toda la talacha, quedandonos:

$$
\int_{x_0}^{x_2}P_2(x)dx=f(x_0)\left[-\frac{(x_0-x_2)}{2}+\frac{(x_0-x_2)^2}{6(x_0-x_1)}\right]+f(x_1)\left[-\frac{(x_0-x_2)^3}{6(x_1-x_0)(x_1-x_2)}\right]+f(x_2)\left[\frac{(x_2-x_1)}{2}-\frac{(x_2-x_1)^2}{6(x_2-x_0)}+\frac{(x_0-x_1)^3}{6(x_2-x_0)(x_2-x_1)}\right]
$$

Remplazando $$x_1=x_0+h, x_2=x_0+2h$$ resulta:
$$
\int_{x_0}^{x_2}P_2(x)dx=f(x_0)\left[\frac{h}{3}\right]+f(x_1)\left[\frac{4h}{3}\right]+f(x_2)\left[\frac{h}{3}\right]$$

Factorizando $\frac{h}{3}$, tenemos:
$$
\int_{x_0}^{x_2}f(x)dx\approx\left[\frac{h}{3}\right] [f(x_0)+4f(x_1)+f(x_2)]
$$

con $h=\frac{b-a}{2}$

Así pues a esta expresión se le conoce como **regla de Simpson**:

$$
\int_{x_0}^{x_2}f(x)dx\approx\left[\frac{b-a}{6}\right] [f(x_0)+4f(x_1)+f(x_2)]
$$

Además esta tiene un error en la aproximación dado por la expresión:

$$
E= \left | \frac{h^5}{90}f^(4)(z)\right | 
$$

con $x_0<z<x_2$.
