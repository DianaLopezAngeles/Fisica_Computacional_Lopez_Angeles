```python
from pylab import *
%matplotlib inline
```

```python
def borrachitow2(L=10,px=[-4,4]):
    trayectoria_x = []
    trayectoria_y = []
    paso_actual= [0,0]
    salida = 0
    while salida == 0:
        volado = rand()
        if volado <=0.25:
            paso_actual[0] += -1
        elif 0.25 < volado <= 0.5:
            paso_actual[0] += 1
        elif 0.5 < volado < 0.75:
            paso_actual[1] += -1
        else:
            paso_actual[1] += 1
        salida = puerta(paso_actual,L,px)
        if salida == 0: 
            paso_actual = corregir_y_checar(paso_actual,L)
        trayectoria_x.append(paso_actual[0])    
        trayectoria_y.append(paso_actual[1])
    return trayectoria_x, trayectoria_y
```

```python
def corregir_y_checar(pos,L):
    if pos[0] == L+1:
        pos[0] += -2
    elif pos[0] == -(L+1):
          pos[0] += 2
    elif pos[1] == L+1:
        pos[1] += -2
    elif pos[1] == -(L+1):
        pos[1] += 2
    return pos

def puerta(pos,L,px):
    salida = 0
    if ( pos[1] == -(L+1) and px[0] < pos[0]< px[1]):
        salida = 1
    return salida

def cantina (L=10,px=[-4,4]):
    xs = [px[1],L,L,-L,-L,px[0]]
    ys = [-L,-L,L,L,-L,-L]
    plot(xs,ys,lw=3)
```

```python
L,px=64,[-4,4]
Lm= L+1
xlim(-Lm,Lm), ylim(-Lm,Lm)
cantina (L,px)
for i in range(1):
    x,y = borrachitow2(L,px)
    plot(x,y), plot(x[-1],y[-1],"o")
show()
```

```python
def prom_trayec(trayectoria):
    N= len(trayectoria)
    promedio = [0,0]
    for i in xrange (N):
        promedio[0] += trayectoria [0][i]
        promedio[1] += trayectoria [1][i]
    promedio[0] = float(promedio[0]/N)
    promedio[1] = float(promedio[1]/N)
    return promedio
```

```python
x,y = borrachitow2(L,px)
print prom_trayec ([x,y])
```
